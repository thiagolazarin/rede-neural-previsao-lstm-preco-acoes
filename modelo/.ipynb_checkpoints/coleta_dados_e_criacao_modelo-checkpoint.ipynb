# Normalizando a classe Close usando o MinMaxScaler:

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_close = scaler.fit_transform(df['Close'])
```

```python
scaled_close.shape
```

```python
plt.figure(figsize=(12,6))
plt.plot(scaled_close, label='Preço Normalizado (Close)')
plt.title('Preço de Fechamento Normalizado - DIS')
plt.xlabel('Tempo')
plt.ylabel('Valor Normalizado')
plt.legend()
plt.grid(True)
plt.show()
```

# Criação das janelas de sequência para LSTM

```python
def create_sequences(data, window_size):
    X, y = [], []
    for i in range(window_size, len(data)):
        X.append(data[i - window_size:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

# definindo tamanho da window
window_size = 60

# gera as sequencias
X, y = create_sequences(scaled_close, window_size)

# reshape para formato [amostras, timesteps, features]
X = X.reshape((X.shape[0], X.shape[1], 1))
```

# Dividindo entre treino e teste

```python
split = int(0.8 * len(X)) 

X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

# Construção do Modelo LSTM:

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dropout(0.2))
model.add(Dense(units=1)) # previsão de um valor
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=32,
    validation_data=(X_test, y_test)
)
```

```python
plt.plot(history.history['loss'], label='Treinamento')
plt.plot(history.history['val_loss'], label='Validação')
plt.title('Perda (Loss) durante o treinamento')
plt.xlabel('Épocas')
plt.ylabel('Erro quadrático médio')
plt.legend()
plt.grid(True)
plt.show()
```

✅ Coisas boas:
Perda de validação é baixa e estável → o modelo generaliza bem.

Perda de treino e validação estão próximas → não há overfitting visível.

A queda rápida nas primeiras épocas é esperada, e depois a curva estabiliza.

🔎 Interpretação:
O modelo aprendeu os padrões dos dados com eficácia.

Com loss < 0.005 na validação, as previsões normalizadas estão muito próximas do valor real.

O uso de Dropout ajudou a evitar overfitting.

# Previsão com o modelo treinado:

```python
# prevendo o X_test com o modelo treinado:
y_pred_scaled = model.predict(X_test)
```

```python
# inverter a normalização. Como usamos apenas o Close para treinar, vamos aplicar inverse_transform para trazer os valores para a escala original:

# y_pred_scaled está no formato (n, 1), então podemos aplicar direto
y_pred = scaler.inverse_transform(y_pred_scaled)

# como y_test também estava normalizado, precisamos inverter ele também
y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python
#3. Calcular métricas de desempenho:
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(y_test_actual, y_pred)
rmse = np.sqrt(mean_squared_error(y_test_actual, y_pred))
mape = np.mean(np.abs((y_test_actual - y_pred) / y_test_actual)) * 100

print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAPE: {mape:.2f}%")
```

```python
# Plotar previsão vs real:

plt.figure(figsize=(12,6))
plt.plot(y_test_actual, label='Valor Real')
plt.plot(y_pred, label='Previsão')
plt.title('Previsão vs Valor Real - Preço de Fechamento')
plt.xlabel('Tempo')
plt.ylabel('Preço (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
dates = df.index[window_size + split:]
plt.figure(figsize=(12,6))
plt.plot(dates, y_test_actual, label='Valor Real')
plt.plot(dates, y_pred, label='Previsão')
plt.title('Previsão vs Valor Real - Preço de Fechamento')
plt.xlabel('Data')
plt.ylabel('Preço (USD)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```
